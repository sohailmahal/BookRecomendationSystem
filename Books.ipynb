## subtituting nan values with zxc in all columns

```python
books_df['subtitle'].fillna('zxc', inplace=True)
books_df['authors'].fillna('zxc', inplace=True)
books_df['categories'].fillna('zxc', inplace=True)
books_df['description'].fillna('zxc', inplace=True)
```

```python
books_df
```

## sort categories column

```python
def sortcategories (data):
    if pd.isna(data)==True:
        pass
    else:
        data = str(data)
        data = data.replace(" ", "")
        data = data.split(',')
        return data
```

```python
books_df['categories'] = books_df['categories'].apply(sortcategories)
books_df['categories']
```

```python
books_df['categories'][25]
```

## sort authour column

```python
def sortauthors (data):
    if pd.isna(data)==True:
        pass
    else:
        data = str(data)
        data = data.replace(" ", "")
        data = data.split(';')
        return data
```

```python
books_df['authors'] = books_df['authors'].apply(sortauthors)
books_df['authors']
```

## sort description

```python
def sortdescription (data):
    if pd.isna(data)==True:
        pass
    else:
        data = str(data)
        data = data.replace("'", "")
        data = data.replace('"', '')
        data = data.replace(":", "")
        data = data.replace(";", "")
        data = data.replace("--", "")
        data = data.replace(",", "")
        data = data.replace(".","" )
        data = data.replace("?","" )
        data = data.replace("(","" )
        data = data.replace(")","" )
        data = data.replace("  ","" )
        data = data.split(' ')
        return data
```

```python
books_df['description'] = books_df['description'].apply(sortdescription)
books_df['description'] 
```

```python
books_df['description'][3] 
```

## sort subtitles

```python
def objtolist(data):
    if pd.isna(data)==True:
        pass
    else:
        data = str(data)
        return data.split(' ')
```

```python
books_df['subtitle'] = books_df['subtitle'] = books_df['subtitle'].apply(objtolist)
books_df['subtitle']
```

```python
books_df
```

```python
books_df['tags'] = books_df['subtitle']+ books_df['authors']+books_df['categories']+books_df['description']
```

```python
books_df['tags'][0]
```

```python
booktag = books_df[['isbn13', 'title', 'tags']]
booktag
```

```python
booktag['tags'] = booktag['tags'].apply(lambda x: ' '.join(x))
booktag['tags'][0]
```

```python
booktag['tags'] = booktag['tags'].apply(lambda x:x.lower())
booktag['tags'][0]
```

```python
booktag['tags'] = booktag['tags'].apply(lambda x: x.replace('zxc', ''))
booktag['tags'][0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
print(ps.stem('run'))
print(ps.stem('running'))
print(ps.stem('runs'))
print(ps.stem('actions'))
```

```python
ps.stem('story')
```

```python
text='sohail was running and performing different actions while dancing'
list1=[]
for i in text.split(' '):
    list1.append(ps.stem(i))

string = ' '.join(list1)
string
```

```python
def portstem (text):
    list1=[]
    
    for i in text.split(' '):
        list1.append(ps.stem(i))
    
    string = ' '.join(list1)
    return string
    
```

```python
booktag['tags'] = booktag['tags'].apply(portstem)
booktag['tags'][0]
```

```python
def bookrecommender (entry):
    index = booktag[booktag['title']==entry].index[0]
    distance = similarity[index]
    
    book_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x:x[1])[1:10]
    
    for i in book_list:
        print(booktag.iloc[i[0]].title)
        
    
```
